```python
from dataset import (
    LeRobotDataset,
    MultiDataset,
    create_dataloader,
)
from dataset import FilterConfig, QAnnotationFilter
from PIL import Image

config = FilterConfig(
    range_nop_path="path/to/RoboInter_Data/range_nop.json",
    q_filters=[
        QAnnotationFilter("Q_annotation.trace", ["Primary"]),
        QAnnotationFilter("Q_annotation.substask", ["Primary"]),
        QAnnotationFilter("Q_annotation.primitive_skill", ["Primary"]),
        QAnnotationFilter("Q_annotation.object_box", ["Primary"]),
        QAnnotationFilter("Q_annotation.placement_proposal", ["Primary"]),
        QAnnotationFilter("Q_annotation.gripper_box", ["Primary"]),
        QAnnotationFilter("Q_annotation.contact_frame", ["Primary"]),
        QAnnotationFilter("Q_annotation.state_affordance", ["Primary"]),
        QAnnotationFilter("Q_annotation.affordance_box", ["Primary"]),
        QAnnotationFilter("Q_annotation.contact_points", ["Primary","Secondary"]),
    ],
    q_filter_mode="all",  # "all" = AND, "any" = OR
)
    
dataloader = create_dataloader(
    "path/to/RoboInter_Data/Annotation_with_action_lerobotv21/lerobot_droid_anno",
    batch_size=32,
    action_horizon=16,
    shuffle=False
)
for batch in dataloader:
    images = batch["observation.images.primary"]  # (B, H, W, C)
    states = batch["state"]                       # (B, 7)
    actions = batch["action"]                     # (B, 16, 7)
    break
print(batch.keys())
print(type(batch["annotation.state_affordance"]))
print(batch["annotation.time_clip"])
print(batch["observation.images.wrist"].shape)
for key in batch.keys():
    if 'action' in key or 'image' in key:
        continue
    print(key)
    print(batch[key][10])
    print(type(batch[key][10]))
img_obj = Image.fromarray(batch["observation.images.primary"][0])
img_obj
```
